##### **Metric Evaluation**
 
|               | Predicted Negative | Predicted Positive |
|---------------|--------------------|--------------------|
| Actual Negative | True Negative (TN)   | False Positive (FP)  |
|                | Karyawan tidak dipromosikan (sesuai prediksi) | Karyawan dipromosikan (kesalahan prediksi) |
| Actual Positive | False Negative (FN)   | True Positive (TP)   |
|                | Karyawan tidak dipromosikan (kesalahan prediksi) | Karyawan dipromosikan (sesuai prediksi) | 

Penjelasan:
- **True Positive (TP)**: Karyawan diprediksi akan dipromosikan dan benar-benar dipromosikan.
- **False Negative (FN)**: Karyawan diprediksi tidak akan dipromosikan padahal seharusnya dipromosikan.
- **False Positive (FP)**: Karyawan diprediksi akan dipromosikan padahal sebenarnya tidak dipromosikan.
- **True Negative (TN)**: Karyawan diprediksi tidak akan dipromosikan dan memang tidak dipromosikan.

Key Metrics:
- **Accuracy**: \((TP + TN) / (TP + TN + FP + FN)\)
- **Precision**: \(TP / (TP + FP)\)
- **Recall (Sensitivity)**: \(TP / (TP + FN)\)
- **F1 Score**: \(2 * (Precision * Recall) / (Precision + Recall)\)

Konsekuensi Bisnis
- False Positive (FP),
karena calon kandidat belum tepat saatnya direkomendasikan untuk promosi jabatan sehingga perusahaan menempatkan karyawan yang belum mumpuni dan berpengaruh pada performa bisnis perusahaan.

- False Negative (FN),
karena calon kandidat yang sudah selayaknya direkomendasikan untuk promosi jabatan tetapi tidak terlaksana sehingga terdapat potensi kehilangan calon kandidat potensial disamping berpotensi tidak terpenuhinya kebutuhan jumlah karyawan yang mengisi saat transisi jabatan.

##### **Multikolinearitas**

```python
# Fungsi untuk mengecek multikolinearitas

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
# Melihat nilai VIF untuk data numerik

calc_vif(numerik.dropna()).sort_values(by='VIF', ascending=False)
```

Dalam multikolinearitas, antar variabel independen tidak boleh ada pengaruh satu dengan lainnya. Dari nilai VIF di atas diatas, terlihat fitur 'age' memiliki nilai multikolineritas yang tinggi sehingga kita nantinya akan mencoba melihat bagaimana performance model apabila kita drop fitur 'age'.

```python
# Balancing Method

# Random
rus = RandomUnderSampler(random_state=42)

# Undersampling Technique
nm1 = NearMiss(version = 1)
nm2 = NearMiss(version = 2)
nm3 = NearMiss(version = 3)

# Oversampling Technique
smote = SMOTE()
```

```python
# benchmarking balancing

skfold = StratifiedKFold(n_splits=5)
balancings = [rus, nm1, nm2, nm3, smote]

score_f1_list = []
score_recall_list = []
score_precision_list = []
std_list = []

for balancing in balancings:
    estimator = Pipeline([
        ('preparation', transformer),
        ('scaling', RobustScaler()),
        ('balancing', balancing),
        ('modelling', xgb)
    ])

    score_f1 = cross_val_score(estimator, X_train, y_train, cv = skfold, scoring='f1')
    score_f1_list.append(score_f1.mean())

    score_recall = cross_val_score(estimator, X_train, y_train, cv = skfold, scoring='recall')
    score_recall_list.append(score_recall.mean())

    score_precision = cross_val_score(estimator, X_train, y_train, cv = skfold, scoring='precision')
    score_precision_list.append(score_precision.mean())

    print(balancing)
    print(score_f1)

    std = cross_val_score(estimator, X_train, y_train, cv = skfold, scoring='f1').std()
    std_list.append(std)

pd.DataFrame({
    'balancing': balancings,
    'mean f1 score': score_f1_list,
    'mean recall score': score_recall_list,
    'mean precision score': score_precision_list,
    'std f1 score': std_list
}).sort_values('mean f1 score', ascending = False)
```

Kita ingin model kita bisa memprediksi kelas positif (True Positif) dengan maksimal. Kita ingin fokus meningkatkan True Positif, dan menekan False Negatif. Sehingga kita ingin nilai recall yang tinggi.
Oleh karena itu, kita pilih untuk melakukan pemodelan dengan menggunakan RandomUnderSampler sebagai metode balancing/resampling.

```python
# Rule based kriteria untuk yang layak dipromosikan

def predict_promotion(rule):
    predictions = []
    
    for index, row in rule.iterrows():
        if row['length_of_service'] < 4:
           predictions.append(0)
        elif row['avg_training_score'] > 80 and row['previous_year_rating'] >= 4:
            predictions.append(1)
        else:
            predictions.append(0)
    return predictions

rule_predictions = predict_promotion(rule)

# Menambahkan kolom prediction
rule['rule_predicted'] = rule_predictions

# Print Classification Report
print('Classification report Rule-Based prediction : ')
print(classification_report(rule['is_promoted'], rule['rule_predicted']))

# Confusion matrix
cm = confusion_matrix(rule['is_promoted'], rule['rule_predicted'])

# Plotting confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Not Promoted', 'Promoted'],
            yticklabels=['Not Promoted', 'Promoted'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## Model Limitation

#### [SHAP](https://shap.readthedocs.io/en/latest/)

```python
# mengubah 'object' features to 'categorical' untuk keperluan SHAP 
object_features = X.select_dtypes(include=['object']).columns
for column in object_features:
    X[column] = X[column].astype('category')

X_train_shap, X_test_shap, y_train_shap, y_test_shap = train_test_split(
    X,
    y,
    test_size = 0.2,
    stratify = y,
    random_state = 1024)
```

```python
import xgboost as xgb

# Instantiate RandomUnderSampler
undersampler = RandomUnderSampler(sampling_strategy=0.85, random_state=42)

# Fit and resample your training data
X_train_resampled, y_train_resampled = undersampler.fit_resample(X_train_shap, y_train_shap)

# Convert data to DMatrix format
dtrain = xgb.DMatrix(data=X_train_resampled, label=y_train_resampled, enable_categorical=True)
dtest = xgb.DMatrix(data=X_test_shap, label=y_test_shap, enable_categorical=True)

# Define parameters for XGBoost
params = {
    'eval_metric': 'logloss',
    'colsample_bytree': 0.6,
    'learning_rate': 0.05,
    'max_depth': 4,
    'n_estimators': 175,
    'subsample': 0.9}

# Train the model
model = xgb.train(
    params=params,
    dtrain=dtrain,
    num_boost_round=5000,
    evals=[(dtest, "test")],
    verbose_eval=100,
    early_stopping_rounds=20)
```

```python
xgb.plot_importance(model, importance_type='gain')
plt.title("xgboost.plot_importance(model, importance_type='gain')")
plt.show()
```

Limitasi model bahwa model kurang dapat melihat dampak yang cukup besar dari fitur gender dan recruitment_channel. Oleh karena itu, untuk ke depannya fitur ini bisa ditiadakan karena dampaknya sangat kecil terhadap model untuk memprediksi target.

```python
# Membuat TreeExplainer untuk model XGBoost
explainer = shap.TreeExplainer(model)

# Menghitung SHAP values
shap_values = explainer.shap_values(dtest)

# Plot summary plot
shap.summary_plot(shap_values, X_test)
```

Gambar di atas menunjukkan nilai SHAP (SHapley Additive exPlanations) dari berbagai fitur dalam model machine learning yang mengukur dampak fitur-fitur tersebut terhadap output model.
